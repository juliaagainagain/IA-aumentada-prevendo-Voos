Analisando a distribuição dos dados

```python
#calcula tamanho ideal dos bins
import numpy as np 
def calcular_largura_bin(df, coluna):
        Q75, Q25 = np.percentile(df[coluna], [75, 25])
        IQR = Q75 - Q25
        largura_bin = 2 * IQR * np.power(len(df[coluna]), -1/3)
        return largura_bin
```

```python
largura_bin_at = calcular_largura_bin (dados, 'arrival_time')
sns.histplot(data= dados, x= 'arrival_time', kde = True, binwidth= largura_bin_at ) #kde faz a linha nos bins, e bins = 10 (ou qualquer outro tamanho) 
#define o tamanho nessa parte existe uma forma de calcular o tamanho exato dos bins mas fiz uma calculadora do tamanho ideal:
```

```python
largura_bin_at = calcular_largura_bin (dados, 'departure_time')
sns.histplot(data= dados, x= 'departure_time', kde = True, binwidth= largura_bin_at )
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import math

def calcular_metricas_regressao(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)  #erro quadratico
    rmse = math.sqrt(mse)  #calcula a raiz quadrada manualmente
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    metricas = {
        'Raiz do Erro Quadrático Médio': round(rmse, 4),
        'Erro Absoluto Médio': round(mae, 4),
        'R2 Score': round(r2, 4)
    }
    return metricas
```
